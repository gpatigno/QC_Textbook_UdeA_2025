```python
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Operator
from qiskit.visualization import array_to_latex
```

```python
# Descomente la siguiente linea si desea ver las im치genes m치s grandes.
%config InlineBackend.figure_format = 'svg' # Permite mejor calidad en las im치genes generadas.
```

```python
q = QuantumRegister(2, name='q')
```

```python
qc = QuantumCircuit(q)
qc.z(q[0]) # Qubit menos significativo en Qiskit
qc.x(q[1]) # Qubit m치s significativo en Qiskit

qc.draw('mpl')
# qc.draw('mpl', style='iqx', filename='XxZCircuit.png')
```

```python
unitary = Operator(qc)
array_to_latex(unitary, prefix="\\text{Circuito } \\text{$X$} \\otimes \\text{$Z$ = }")
```

```python
qc = QuantumCircuit(q)
qc.x(q[0]) # Qubit menos significativo en Qiskit
qc.x(q[1]) # Qubit m치s significativo en Qiskit

qc.draw('mpl')
# qc.draw('mpl', style='iqx', filename='X-X.png')
```

```python
unitary = Operator(qc)
array_to_latex(unitary, prefix="\\text{Circuito } \\text{$X$} \\otimes \\text{$X$ = }")
```
