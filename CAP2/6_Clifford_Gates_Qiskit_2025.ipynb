```python
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Operator
from qiskit.visualization import array_to_latex
```

```python
# Descomente la siguiente linea si desea ver las imágenes más grandes.
%config InlineBackend.figure_format = 'svg' # Permite mejor calidad en las imágenes generadas.
```

## Proving Clifford Gates Conjugation:

```python
q = QuantumRegister(1, name='q')
qc = QuantumCircuit(q)

qc.h(q[0])
qc.x(q[0])
qc.h(q[0])

qc.draw('mpl')
# qc.draw('mpl', style='iqx', filename='HXHConjugating.png')
```

```python
unitary = Operator(qc)
array_to_latex(unitary, prefix="\\text{$H \\cdot X \\cdot H$ = $HXH$ =  }\n")
```

```python
q = QuantumRegister(2, name='q')
qc = QuantumCircuit(q)

qc.cx(q[0],q[1])
qc.x(q[0])
qc.cx(q[0],q[1])

qc.draw('mpl')
# qc.draw('mpl', style='iqx', filename='CX-X-CX_Conjugate.png')
```

```python
unitary = Operator(qc)
array_to_latex(unitary, prefix="\\text{$CX (X \\otimes I) CX$ = }\n")
```

#### Which must be equivalent to the following circuit:

```python
qc = QuantumCircuit(q)
qc.x(q[0])
qc.x(q[1])

qc.draw('mpl')
# qc.draw('mpl', style='iqx', filename='X-X.png')
```

```python
unitary = Operator(qc)
array_to_latex(unitary, prefix="\\text{$X \\otimes X$ = }\n")
```

#### Confirming the equivalence by Conjugation.
