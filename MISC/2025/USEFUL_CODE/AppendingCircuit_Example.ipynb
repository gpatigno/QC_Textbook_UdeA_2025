```python
from qiskit import QuantumCircuit

# Define a simple 1-qubit quantum circuit
qc1 = QuantumCircuit(1, name="Hadamard-X")  # Naming the small circuit
qc1.h(0)  # Apply Hadamard gate
qc1.x(0)  # Apply X (Pauli-X) gate

# Convert it to an instruction with a label
qc1_inst = qc1.to_instruction()

# Define a 3-qubit quantum circuit with classical registers
qc3 = QuantumCircuit(3, 3)

# Append the named 1-qubit circuit to qubit 1 in the 3-qubit circuit
qc3.append(qc1_inst, [1])

# Add a measurement for all qubits
qc3.measure([0, 1, 2], [0, 1, 2])

# Draw the circuit
qc3.draw('mpl')
```
