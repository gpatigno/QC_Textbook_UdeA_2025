```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator, Statevector, DensityMatrix

qc = QuantumCircuit(1)
qc.h(0)
```

```python

print(Operator(qc).to_matrix()) # unitary
```

```python
print(Statevector(qc))
```

```python

print(Statevector(qc).probabilities())
```

```python
print(Statevector(qc).to_dict())
```

```python
print(DensityMatrix(qc))
```

```python
print(DensityMatrix(qc).to_operator().to_matrix())
```

```python
type(Statevector(qc))
```

```python
type(DensityMatrix(qc).to_operator().to_matrix())
```

```python
from qiskit.visualization import array_to_latex
```

```python
array_to_latex(Statevector(qc), prefix="\\text{Statevector} = ")
```

```python
array_to_latex(DensityMatrix(qc), prefix="\\text{Circuit = }\n")
```

```python
from qiskit.visualization import plot_bloch_multivector
```

```python
plot_bloch_multivector(Statevector(qc))
```
